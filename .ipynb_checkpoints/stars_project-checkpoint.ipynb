__INTRODUCTION__

A galaxy’s evolution, history and dynamics can be traced by the composition of its stars. In this project, I utilized classification techniques in class to predict the star type through a tuned classification model using 2 predictors, Absolute Magnitude and Relative Radius of the star. Stars can be classified based on many factors such as Temperature, Luminosity, Radius, Absolute Magnitude, and Colour. There are 6 types of stars (Type 0-6) that correspond to Red Dwarf (0), Brown Dwarf (1), White Dwarf (2), Main Sequence (3), Super Giants (4) and Hyper Giants (5), respectively. Upon wrangling and visualizing the different predictors with the Type class, it appears that the absolute magnitude and relative radius of a star correlates best with the 6 types of stars. Absolute magnitude describes the absolute intrinsic brightness of a star based on a specific distance to the star from an observed point (Strobel, 2022), and the Relative Radius of a star describes the radius of the star relative to the radius of the Sun (Ryden, 2003). Based on my data analysis, I found that the Absolute Magnitude of stars decreased from type 0-5, with Red Dwarf being the brightest and Hyper Giants being in the dimmest; whereas the Relative Radius of stars generally increased from type 0-5, with white dwarf stars being the smallest and hyper giants being the largest. __My predictive question will be if we can use classification techniques (K nearest neighbors) to predict the type of star when we are given an Absolute Magnitude and Relative Radius value.__ The stars.csv file was retrieved from Kaggle which was referenced from NASA reports and Wikipedia.

```r
stars_subset<- stars %>%
    select(Temperature, L, R, A_M, Type)

names<- colnames(stars_subset |>
                select(-Type))

stars_subset
```

__Figure 4: Stars Subset Table with Select Columns Relevant to Project__

I created a subset of the stars dataset to just the columns that were relevant to the project to decide which predictors to use (i.e Temperature, L, R, A_M, and Type).

```r
stars_formula <- paste("Type", "~", paste(names, collapse="+"))
stars_formula
```

```r

accuracies <- tibble(size = integer(), 
                     model_string = character(), 
                     accuracy = numeric())

knn_spec <- nearest_neighbor(weight_func = "rectangular", 
                             neighbors = tune()) |>
     set_engine("kknn") |>
     set_mode("classification")


stars_vfold <- vfold_cv(stars_subset, v = 5, strata = Type)

n_total <- length(names)

selected <- c()

# for every size from 1 to the total number of predictors
for (i in 1:n_total) {
    # for every predictor still not added yet
    accs <- list()
    models <- list()
    for (j in 1:length(names)) {
        # create a model string for this combination of predictors
        preds_new <- c(selected, names[[j]])
        model_string <- paste("Type", "~", paste(preds_new, collapse="+"))

        # create a recipe from the model string
        stars_recipe <- recipe(as.formula(model_string), 
                                data = stars_subset) |>
                          step_scale(all_predictors()) |>
                          step_center(all_predictors())

        # tune the KNN classifier with these predictors, 
        # and collect the accuracy for the best K
        acc <- workflow() |>
          add_recipe(stars_recipe) |>
          add_model(knn_spec) |>
          tune_grid(resamples = stars_vfold, grid = 10) |>
          collect_metrics() |>
          filter(.metric == "accuracy") |>
          summarize(mx = max(mean))
        acc <- acc$mx |> unlist()

        # add this result to the dataframe
        accs[[j]] <- acc
        models[[j]] <- model_string
    }
    jstar <- which.max(unlist(accs))
    accuracies <- accuracies |> 
      add_row(size = i, 
              model_string = models[[jstar]], 
              accuracy = accs[[jstar]])
    selected <- c(selected, names[[jstar]])
    names <- names[-jstar]
}
accuracies
```

__Figure 8__ 

I visualized Radius and A_M on a facet grid with the Type column as the colour to analyze the different distributions of the star values.

```r
set.seed(200)

stars_split <- initial_split(stars_subset_final, prop= 0.75, strata= Type)
stars_train <- training(stars_split)
stars_test <- testing(stars_split)
#splitting data into training and test set
```

```r
set.seed(200)

stars_recipe <- recipe(Type ~ A_M + R, data= stars_train) |>
    step_scale(all_predictors()) |>
    step_center(all_predictors())

training_vfold<- vfold_cv(stars_train, v= 5, strata= Type)
training_recipe <- recipe(Type~ A_M + R, data= stars_train)

knn_tune<- nearest_neighbor(weight_func= "rectangular", neighbors = tune()) |>
    set_engine("kknn") |>
    set_mode("classification")

training_resample_fit <- workflow() |>
    add_recipe(training_recipe) |>
    add_model(knn_tune) |>
    tune_grid(resamples = training_vfold, grid = 10)

training_metrics <- collect_metrics(training_resample_fit)

accuracies <- training_metrics |> 
    filter(.metric == "accuracy")

cross_val_plot<- ggplot(accuracies, aes(x = neighbors, y = mean))+
       geom_point() +
       geom_line() +
       labs(x = "Neighbors", y = "Accuracy Estimate") +
        ggtitle("Figure 9: Accuracy Estimate vs Neighbors Plot")+
       scale_x_continuous(breaks = seq(0, 14, by = 1)) +  # adjusting the x-axis
       scale_y_continuous(limits = c(0.4, 1.0)) # adjusting the y-axis

cross_val_plot
```

```r
set.seed(200)

testing_recipe <- recipe(Type~ A_M + R, data= stars_test) |>
                        step_scale(all_predictors()) |>
                        step_center(all_predictors())

knn_spec <- nearest_neighbor(weight_func = "rectangular", neighbors= 2) |>
    set_engine("kknn") |>
    set_mode("classification")

stars_data_fit <- workflow() |>
    add_recipe(testing_recipe) |>
    add_model(knn_spec) |>
    fit(data= stars_train) 

stars_data_fit

stars_test_predictions <- predict(stars_data_fit, stars_test) |>
    bind_cols(stars_test)
stars_test_predictions

stars_test_predictions |> 
metrics(truth= Type, estimate= .pred_class) |>
    select(.metric, .estimate)|>
    head(1)

stars_test_predictions |>  
conf_mat(truth= Type, estimate= .pred_class)
```

__Figure 9__

I then began to tune my classification model, by first setting a seet to 200 to ensure reproducibility and splitting my stars dataset to 75% for training and 25% for testing, with classification based on Type. My recipe was coded to predict Type based on Absolute Magnitude and Radius, and all predictors were scaled and centered. I then tuned my model using cross validation with 10 folds to find that k=2 neighbors yielded the highest accuracy estimate. I then refit a knn model with k=2 onto my testing data and created a column entitled .pred_class for what my model predicted. I also created a confusion matrix and selected for the accuracy yield, which came out to be 82% with 11 incorrect predictions.
